## 01-1 플라스크 시작하기

* Flask는 2004년 오스트리아의 오픈소스 개발자 `아르민 로나허 Armin Rohacher가 만든 웹프레임워크`
* 플라스크는 아르민 로나허가 만우절에 장난삼아 던진 아이디어 였는데 관심이 높아져 서비스로 만들었다고 한다.
* `Flask는 Django와 더불어 파이썬 웹 프레임워크의 양대 산맥`으로 자리매김하고 있다.

><b>플라스크는 마이크로 웹프레임워크이다.</b>
>* Flask는 `마이크로 웹프레임워크`라고 부른다. 여기서 마이크로는 `한 개의 파이썬 파일로 작성`할 수 있다 
>* 또는 기능이 부족하다의미가 아니라 `프레임워크를 간결하게 유지하고 확장할 수 있도록 만들었다는 뜻`이다.

><b>확장성 있는 설계란?</b>
>* Flask에는 폼 Form, 데이터베이스 DataBase를 처리하는 기능이 없다.
>* Django는 폼과 데이터베이스를 처리하는 기능을 포함하고 있다. 쉽게 말해 덩치가 큰 프레임워크이다.
>* Flask는 확장 모듈이라는 것을 사용하여 이를 보완한다.
>* 이 의미는 처음부터 모든 기능을 포함하고 있지 않고 개발자가 필요에 따라 확장 모듈을 포함시켜가며 개발하면 된다.

><b>플라스크는 자유로운 프레임워크다.</b>
>* Flask는 한마디로 자유도가 높은 프레임워크이다.
>* 다른 프레이워크와는 달리 플라스크는 최소한의 규칙만 있기 때문에 개발의 자유도는 다른 프레임워크보다 높다.

### 3. ORM적용

* init.py파일 수정해서 SQLAlchemy를 적용
  - app.config.from_object(config) : config.py정보를 호딩
  - 전역변수로 db, migrate객체를 생성
  - db, migrate객체를 지역변수로 사용하면 블루프린트 같은 다른 모듈에서 사용할 수 없다. 
  - init_app메서드를 이용해서 객체를 초기화

```python
from flask import Flask
from flask_migrate import Migrate
from flask_sqlalchemy import SQLAlchemy
import config
    
db = SQLAlchemy()
migrate = Migrate()

def create_app():
    app = Flask(__name__)
    app.config.from_object(config)
    
    # ORM
    db.init_app(app)
    migrate.init_app(app, db)
    
    # blueprint
    from .views import main_views
    app.register_blueprint(main_views.bp)    
    
    return app
```

#### 2. 모델생성하기

* models.py
  - `db는 init.py에서 생성한 SQLAlchey객채`
  - db.relationship은 참조할 모델(Question)을 정의하고 backref는 역참조설정
    - 역참조설정이란 Question에서 Answer를 참조하는 것을 의미
    - 질문에 여러개의 답변이 달릴 수 있는데 역참조는 이 질문에 달린 답변을 참조할 수 있게 한다.
    - 질문에 해당하는 객체가 a_question이라면 a_question.answer_set와 같은 코드로 해당질문의 답변을 참조

```python
from pybo import db

class Question(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    subject = db.Column(db.String(200), nullable=False)    
    content = db.Column(db.Text(), nullable=False)   
    create_date = db.Column(db.DateTime(), nullable=False)  
    
class Answer(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    question_id = db.Column(db.Integer, db.ForeignKey('question.id', ondelete='CASCADE'))
    question = db.relationship('Question', backref=db.backref('answer_set'))
    content = db.Column(db.Text(), nullable=False)
    create_date = db.Column(db.DateTime(), nullable=False)     
```

#### 1.2 답변 블루프린트 만들기

1. answer_view.py

```python
from datetime import datetime

from flask import Blueprint, url_for, request
from werkzeug.utils import redirect

from pybo import db
from pybo.models import Question, Answer

bp = Blueprint('answer', __name__, url_prefix='/answer') # answer_vewis.py파일이 answer이라는 이름의 BP파일임을 나타냄

@bp.route('/create/<int:question_id>', methods=('POST',)) # 답변저장템플릿의 form태그의 method
def create(question_id):
    question = Question.query.get_or_404(question_id)
    content = request.form['content']  # name속성이 content인 값
    answer = Answer(content=content, create_date=datetime.now())
    question.answer_set.append(answer) # 질문에 대한 답변들
    db.session.commit()
    return redirect(url_for('question.detail', question_id=question_id))
```

* create함수의 question_id는 URL에 전달된다.` http://localhost:5000/answer/create/1/를 요청 받으면 question_id에는 1이 전달`된다
* @bp.route('/create/<int:question_id>', methods=('POST',))에 post방식이 아니라면 아래와 같은 에러가 표시된다.

<img src="./img/02.플라스크 개발 기초_08_methodl_error.png"/>

* request.form['content']는 post방식으로 전송된 항목중 name속성이 content태그의 값을 리턴한다.
* question.answer_set는 질문에 달린 답변들을 의미한다.
  - Question과 Answer모델이 연결되어 있어 backref에 설정한 answer_set을 사용할 수 있다는 것을 기억하자
* `redirect의 question_id는 question_views.py파일에 있는 detail함수의 매개변수로 전달`된다.

    - 참고 : https://wtforms.readthedocs.io/en/2.3.x/validators/#built-in-validators

```python
from datetime import datetime

from flask import Blueprint, render_template, request, url_for
from werkzeug.utils import redirect
from .. import db

from pybo.models import Question
from ..forms import QuestionForm

bp = Blueprint('question', __name__, url_prefix="/question")

@bp.route('/list/')
def _list():
    question_list = Question.query.order_by(Question.create_date.desc())
    return render_template('question/question_list.html', question_list=question_list)

@bp.route('/detail/<int:question_id>/')
def detail(question_id):
    # question = Question.query.get(question_id)
    question = Question.query.get_or_404(question_id)
    return render_template('question/question_detail.html', question=question) 

@bp.route('/create/', methods=('GET', 'POST'))
def create():
    form = QuestionForm()
    if request.method == 'POST' and form.validate_on_submit():
        question = Question(subject=form.subject.data, content=form.content.data, 
            create_date=datetime.now())
        db.session.add(question)
        db.session.commit()
        return redirect(url_for('main.index'))    
    return render_template('question/question_form.html', form=form)
```

* form.validate_on_submit 함수는 전송된 폼 데이터의 정합성을 점검 
* 즉, DataRequired() 같은 점검 항목에 이상이 없는지 확인
* 즉, 위 코드에 추가한 내용은 POST 요청이고 폼 데이터에 이상이 없을 경우 
* 질문을 저장한 뒤 main.index 페이지로 이동하라는 내용
* 추가한 코드의 핵심은 데이터 전송 방식이 POST인지 GET인지에 따라서 달리 처리하는 부분이다.
