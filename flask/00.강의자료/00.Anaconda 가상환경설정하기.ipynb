### Anaconda 가상환경

env는 Python의 독립적인 가상의 실행 환경(Environment)를 말합니다.
아나콘다 패키지로 설치한 경우 conda 명령어를 이용해서 env를 생성/삭제 및 관리할 수 있습니다.

#### env 생성
>conda create -n snowdeer_env python=3.5<br>
>conda create -n snowdeer_env python=3.6.5 tensorflow keras
* -n or -name

* Python 3.5 버전의 ‘snowdeer_env’라는 이름으로 env를 생성합니다. (현재 Windows에서 Keras를 설치할 거면 Python 버전을 3.6이 아닌 3.5로 해야 합니다.)

#### env 리스트 보기
>conda env list
* or conda info -e


#### env 활성화 / 비활성화
>activate snowdeer_env<br>
>deactivate<br>

#### env 삭제
>conda env remove -n snowdeer_env


#### env에 Tensorflow, Keras 설치
>activate snowdeer_env<br>
>pip install tensorflow<br>
>pip install keras<br>

#### Tensorflow와 Keras의 설치 확인은 다음 명령어를 이용해서 확인할 수 있습니다.

###### Tensorflow-gpu
* https://kamang-it.tistory.com/entry/Tensorflow텐서플로우-설치하기 

1. cuda설치
   - https://developer.nvidia.com/cuda-downloads 에가서 쿠다를 설치한다.
   - c:\Program Files\NVIDIA GPU Computing Toolkit\CUDA
   - 이경로는 중요한데 왜냐하면 cudnn을 설치해야하기 때문이다.
2. cudnn설치
   - https://developer.nvidia.com/rdp/cudnn-download
   - 3 개의 폴더가 존재한다. 이 폴더를 각각 cuda의 설치경로에 동일이름안에 넣어주면 준비는 끝났다.
   
3. tensorflow 사용하기
   - conda create -n <환경명> python=<버전(ex:3.5이나 2.7 등)>
   - conda env list
   - activate <환경명> or source activate <환경명>
   - activate시킨후에 pip로 설치하면 된다.
     - gpu 미사용 -  pip install tensorflow
     - gpu 사용 - pip install tensorflow-gpu
     - 설치후 python에 들어가서 import tensorflow를 했을때 아무 오류가 나지 않으면 성공
     - 참고로 gpu 사용버전이나 미사용버전이나 import tensorflow로 명령어는 똑같다.
