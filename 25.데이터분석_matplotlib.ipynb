## matplotlib 옵션

* http://matplotlib.org/examples/color/named_colors.html

```python
"""
========================
Visualizing named colors
========================

Simple plot example with the named colors and its visual representation.
"""
from __future__ import division

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(8, 5))

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

### Figuer객체

* 모든 그래프에는 `Figure객체` 정식으로는 `matplotlib.figure.Figure`클래스 객체가 포함되어 있다.
* 내부 그래프가 아닌 경우에는 하나의 figure에는 하나의 window를 가진다.
* Jupyter Notebook에서는 윈도우객체가 생성되지 않지만 독립적으로 실행할 경우에는 하나의 Figure당 하나의 id(메모리주소)와 윈도우를 별도로 가진다.
* 원래는 Figure를 생성하려면 figure라는 명령을 사용해서 Figure객체를 생성하지만 일반적 plot명령을 실행하면 Figure객체를 생성해주기 때문에 figure명령을 잘 사용하지는 않는다.
* figure명령을 명시적으로 사용하는 경우에는 `여러개의 윈도우를 동시`에 띄우거나 Jupyter Notebook에서는 `figure의 크기를 설정`할 경우에 사용한다.

```python
# 그림크기를 설정하기
np.random.seed(42)
# plt.figure?
plt.figure(figsize=[20, 2])  # figsize는 tuple or list로 설정가능
plt.title('그림크기설정하기')
plt.plot(np.random.randn(200))
plt.show()
```

### Axes객체와 subplot명령

* 하나의 Figure객체안에 여러개의 그래프를 배열형태로 나타낼 경우에는 Figure객에안에 `Axes`라고 하는 객체가 있다.
* Figure객체안에 서브객체인 `Axes를 생성하려면 subplot`명령을 사용해서 명시적으로 Axes객체를 생성해야 한다.
* 하지만, plot명령을 사용하면 자동으로 Axes객체를 생성해 준다.
* `subplot명령은 그리드(grid)형태의 Axes객체를 생성`하는데 Figure가 행렬(matrix)이고 Axes가 행렬의 요소라고 생각하면 된다.
* 주의할 점은 첫 번째 Axes의 순서가 0이 아니라 1부터 시작된다.

```python
# 1. subplot() : 2행 1열
# 2개의 그래프를 그릴 경우에 행이 2, 열이1 즉 2x1인 행렬이다.
# subplot명령은 세개의 인수값을 가지는 데 (행, 열, 순서)형태로 정의한다.
# subplot(2,1,1) or subpblot(2,1,2) -> subplot(211), subplot(212)
plt.figure(figsize=[10, 3]) 

x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)

x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1, y1, 'r-*')

plt.subplot(212)
plt.plot(x2, y2, 'g:o')

plt.tight_layout() # 그래프간의 간격을 자동으로 정렬
plt.show()
```

```python
# 2. subplot 2행 2열
plt.figure(figsize=[10, 3]) 

plt.subplot(221)
plt.title('1st axes - 221')
plt.plot(np.random.rand(5))

plt.subplot(222)
plt.title('2nd axes - 222')
plt.plot(np.random.randn(5))

plt.subplot(223)
plt.title('3rd axes - 223')
plt.plot(np.random.randn(5))

plt.subplot(224)
plt.title('4th axes - 224')
plt.plot(np.random.randn(5))

plt.tight_layout()
plt.show()
```

```python
# 3. subplot으로 1행 3열
# 1st-bar, 2nd-scatter, 3-line
names = ['손흥민', '이강인', '김민재']
values = [50, 60, 100]

plt.figure(figsize=[10, 3]) 

plt.subplot(131)
plt.title('1st axes - bar')
plt.bar(names, values)

plt.subplot(132)
plt.title(' 2nd axes - scatter')
plt.scatter(names, values)

plt.subplot(133)
plt.title('3rd axes - line')
plt.plot(names, values)

plt.suptitle('전체그래프제목')
plt.tight_layout()
plt.show()
```

#### 6. 상자그래프(box)

* 기술통계학에서 box plot은 `수치적 통계자료(기술통계)를 표현`하는 그래프이다.
* 이 그래프에서는 자료를 기준으로 `5개의 요약수치`를 기본으로 그래프를 그린다.
  1. 최소값    : 1사분위수에서 1.5IQR값을 뺀 위치
  1. 1사분위수 : 25%위치
  1. 2사분위수 : 50%위치로 중앙값(median)을 의미
  1. 3사분위수 : 75%위치 
  1. 최대값    : 3사분위수 1.5IQR를 더한 위치
  1. 이상치    : 최소값/최대값을 벗어나는 위치

```python
matplotlib.rcParams["figure.figsize"]= [10,6]

import matplotlib.colors as mcolors
from numpy.random import multivariate_normal

data = np.vstack([
    multivariate_normal([10, 10], [[3, 2], [2, 3]], size=100000),
    multivariate_normal([30, 20], [[2, 3], [1, 3]], size=1000)
])

gammas = [0.8, 0.5, 0.3]

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].set_title('Linear normalization')
axes[0, 0].hist2d(data[:, 0], data[:, 1], bins=100)

for ax, gamma in zip(axes.flat[1:], gammas):
    ax.set_title(r'Power law $(\gamma=%1.1f)$' % gamma)
    ax.hist2d(data[:, 0], data[:, 1],
              bins=100, norm=mcolors.PowerNorm(gamma))

fig.tight_layout()

plt.show()
```
