```python
# 1. numpy : 난수발생
print(dir(np.random))
print()

data = np.random.rand() # 0<= random < 1
print(type(data), data)

data = np.random.rand(50)
print(type(data), data)
```

```python
# 2. pandas

# 1) 1차원 자료 생성
# pd.Series?
s = pd.Series(data)
print(type(s), s)
print(dir(s))
```

```python
# pandas의 기능을 이용한 시각화
s.plot()
s.plot?
```

```python
# 2) 2차원 자료 생성
# pd.DataFrame?

# numpy로 랜덤하게 10행 3열의 2차원배열(정확하게는 행렬, matrix)
data = np.random.rand(10,3)
print(type(data), data)
```

```python
# 3) 2차원 자료구조 : pd.DataFrame
df = pd.DataFrame(data)
print(type(df), df)
display(df)
```

```python
# 컬럼명정의하기
df = pd.DataFrame(data, columns=[2019, '2020', 2021])
df
```

```python
# 4) 시각화 : pandas로 2차원 그래프 그리기(기본 line 그래프)
df.plot()
```

```python
df.plot(kind='bar')
```

```python
# 수평바
df.plot(kind='barh')
```

```python
# area
df.plot(kind='area') # 누적그래프 : 기본값 stacked=True
df.plot(kind='area', stacked=False) # 비누적 
```

```python
# 파이차트
# 5개의 난수를 생성, Series자료형, pie챠트로 시각화
np.random.seed(42)
ser = pd.Series(data=np.random.rand(5), index=['2019','2020','2021','2022','2023'])
print(ser)

# 시각화
# autopct : 소숫점 2자리로 설정
ser.plot(kind='pie', autopct="%.2f%%", figsize=(10,5), fontsize=8)
```
